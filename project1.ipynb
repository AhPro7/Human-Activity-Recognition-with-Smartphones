 ## Models


1_LinearDiscriminantAnalysis

````python
x = train_data.drop('Activity',axis=1)
y = train_data['Activity']

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=145)
````

````python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
LDA = LinearDiscriminantAnalysis()
LDA.fit(x_train,y_train)
````

````python
yy=test_data['Activity']
xx=test_data.drop('Activity',axis=1)

predict =  LDA.predict(xx)
````

````python
from sklearn.metrics import classification_report,accuracy_score
print(classification_report(predict,yy))
````

````python
print(accuracy_score(predict,yy))
````

2_LogisticRegression


````python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression(random_state=110)
log.fit(x_train,y_train)
````

````python
predict0 =  log.predict(xx)
````

````python
print(classification_report(predict0,yy))
````

````python
print(accuracy_score(predict0,yy))
````
